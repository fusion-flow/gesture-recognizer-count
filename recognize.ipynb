```python
import cv2
import mediapipe as mp
import time
import os
```

```python
images = []
folder_path = "fingers"
for filename in os.listdir(folder_path):
    img_ = cv2.imread(os.path.join(folder_path,filename))
    if img_ is not None:
        images.append(img_)


mpHands = mp.solutions.hands
hands = mpHands.Hands()
mpDraw = mp.solutions.drawing_utils
fingerCoordinates = [(8,6), (12,10), (16, 14), (20,18)]
thumbCoordinate= (4,2)

```

```python
pTime = 0
cTime = 0
i = 0
print("running..")
while i<1:
    print(i)
    
    img = images[i]
    imgRGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    # show image
    # cv2.imshow("Finger counter", img)
    # # wait untill user presses a key
    # cv2.waitKey(0)
    # cv2.destroyAllWindows()

    results = hands.process(imgRGB)
    multiLandmarks = results.multi_hand_landmarks
    if results.multi_handedness:
            label = results.multi_handedness[0].classification[0].label  # label gives if hand is left or right
            #account for inversion in webcams
            if label == "Left":
                label = "Right"
            elif label == "Right":
                label = "Left"

    if multiLandmarks:
        print("processing landmarks..")
        handPoints = []
        for handLms in multiLandmarks:
            mpDraw.draw_landmarks(img, handLms, mpHands.HAND_CONNECTIONS)
        
            for idx, lm in enumerate(handLms.landmark):
                # print(idx, lm)
                height, width, colorChannel = img.shape
                cx, cy = int(lm.x*width), int(lm.y*height)

                handPoints.append((cx, cy)) # save all the landmarks


        for point in handPoints:
            # print(point)
            # cv2.circle(img, point, 10, (0,0,255), cv2.FILLED)
            # circle with point and number
            cv2.circle(img, point, 10, (0,0,255), cv2.FILLED)
            cv2.putText(img, str(handPoints.index(point)), point, cv2.FONT_HERSHEY_PLAIN, 1, (255,0,0), 1)

        
        # print("Hand points: ")
        # print(handPoints)
        # count number of hands
        upCount = 0
        for coordinates in fingerCoordinates:
            print("Proposed coordinates: ", coordinates)
            if handPoints[coordinates[0]][1] < handPoints[coordinates[1]][1]:
                print("up coordinate: ", coordinates, upCount)
                upCount += 1

        if handPoints[thumbCoordinate[0]][0] < handPoints[thumbCoordinate[0]][1]:
            print("coordinates: ", handPoints[thumbCoordinate[0]][1], handPoints[thumbCoordinate[1]][1])
            print("up coordinate thumb: ", thumbCoordinate, handPoints)
            upCount += 1
        
        cv2.putText(img, str(upCount), (150,150), cv2.FONT_HERSHEY_PLAIN, 12, (255,0,0), 12)

    cv2.imshow("Finger counter", img)
    # wait untill user presses a key
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    i += 1
```
